```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
# had to delete starter api file because it would not read into code even after saving, only showed 'YOUR KEY HERE!'
from url_keys import url_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# define query url
temp_units = "imperial"
query_url = f"http://api.openweathermap.org/data/2.5/weather?appid={url_key}&units={temp_units}&q="
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
cities_check = []
cities_id = []
cloud = []
country = []
dates = []
humid = []
lats = []
lngs = []
maxt = []
mint = []
wind = []

# from github examples 
count_one = 0
set_one = 1

print(f"-\nAnd Now the Weather:\n-")
for city in cities:
    try:
        response = requests.get(query_url + city.replace(" ","&")).json()
        cities_id.append(response['id'])
        cloud.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        dates.append(response['dt'])
        humid.append(response['main']['humidity'])
        lats.append(response['coord']['lat'])
        lngs.append(response['coord']['lon'])
        maxt.append(response['main']['temp_max'])
        mint.append(response['main']['temp_min'])
        wind.append(response['wind']['speed'])
        if count_one > 48:
            count_one = 1
            set_one += 1
            cities_check.append(city)
        else:
            count_one += 1
            cities_check.append(city)
        print(f"Processing {count_one} of Set {set_one} | {city}")
    except Exception:
        print("City 404. Thank you, next.")
print(f"---\nRetrieval Complete:\n---")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create dataframe
weather_df = pd.DataFrame({"City Name":cities_check,
                           "City ID":cities_id,
                          "Cloudiness":cloud,
                         "Date":dates,
                         "Humidity":humid,
                         "Latitude":lats,
                         "Longitude":lngs,
                         "Max Temp":maxt,
                         "Min Temp":mint,
                         "Wind Speed":wind})
weather_df.head()
```

```python
weather_df.to_csv(r'..\06_Python_APIs\Weather Dataframe.csv')
```

#### Latitude vs. Temperature Plot

```python
plt.scatter(weather_df["Latitude"],weather_df["Max Temp"],edgecolors="black",facecolors="blue")
plt.title("City Latitude vs. Max Temperature (10/01/2019)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.grid(b=True,which="major",axis="both",linestyle="-",color="lightgrey")
plt.savefig("Lat_v_Tem.png")
plt.show()
```

```python
plt.scatter(weather_df["Latitude"],weather_df["Humidity"],edgecolors="black",facecolors="blue")
plt.title("City Latitude vs. Humidity (10/01/19)")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.ylim(10,110)
plt.grid (b=True,which="major",axis="both",linestyle="-",color="lightgrey")
plt.savefig("Lat_v_Hum.png")
plt.show()
```

#### Latitude vs. Cloudiness Plot

#### Latitude vs. Wind Speed Plot

```python
plt.scatter(weather_df["Latitude"],weather_df["Wind Speed"],edgecolors="black",facecolors="blue")
plt.title("City Latitude vs. Wind Speed (10/01/19)")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.ylim(0,35)
plt.grid (b=True,which="major",axis="both",linestyle="-",color="lightgrey")
plt.savefig("Lat_v_Win.png")
plt.show()
```

Observable Trends:
1. Max temperatures at the southernmost latitudes are higher than max temperatures at the northernmost latitudes. I expect this is connected to Earth's axial tilt during the dataset's time of year. 
2. City humidity appears to increase closer to the equator. Upper and lower latitudes vary in humidity. 
3. Cloudiness appears to have no strong relationship to latitude. However, the grouping of cities nearly every 20% increase is interesting. 
4. Wind speeds cluster below 20mph for most listed cities regardless of latitude.
